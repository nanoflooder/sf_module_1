# 35. Какие актеры чаще всего снимаются в одном фильме вместе
Варианты ответа:
1. Johnny Depp & Helena Bonham Carter
2. Hugh Jackman & Ian McKellen
3. Vin Diesel & Paul Walker
4. Adam Sandler & Kevin James
5. Daniel Radcliffe & Rupert Grint

```python
# делаем список актёров
unique_actors=actors['actor'].unique()
# делаем сводную таблицу актёр-актёр
pivot=pd.DataFrame(columns=unique_actors,index=unique_actors,dtype='int64').fillna(0)
# убираем из data всё лишнее для ускорения
df=data[['imdb_id','cast']]

# проходим по каждому актёру в списке
for i in range(0,len(unique_actors)):
    # выбираем все фильмы с ним
    df_thisactor=df[df['cast'].str.contains(unique_actors[i])]
    # раскрываем разделители "|" в отдельные строки
    df_thisactor_full=pd.DataFrame(df_thisactor.cast.str.split('|').tolist()).stack().reset_index()[[0]]
    # убираем строки с ним (каждый актёр больше всего снимался в фильмах сам с собой)
    df_thisactor_full=df_thisactor_full[df_thisactor_full[0]!=unique_actors[i]].reset_index()
    # проходим по получившемуся списку других актёров (которые снимались с ним)
    for j in range(0,len(df_thisactor_full.index)):
        # увеличиваем соответствующую ячейку сводной таблицы на 1
        pivot[unique_actors[i]][df_thisactor_full[0][j]]+=1

# сводную таблицу превращаем в обычную
result=pivot.unstack().reset_index()
# для сортировки надо обозвать столбцы
result.columns=['actor1','actor2','cnt']
# убираем все строки с нулями и сортируем по количеству
result=result[result['cnt']>0].sort_values(by='cnt',ascending=False)

display(result.head())
answer_ls.append(5)
```
